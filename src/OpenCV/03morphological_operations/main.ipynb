```python
import cv2
import numpy as np
from IPython import display
from matplotlib import pyplot as plt
```

```python
def cv2_show(img: np.array, name: str = 'img'):
    cv2.imshow(name, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

## 腐蚀操作

```python
img = cv2.imread("../../../data/OpenCV-img/shape.png")
display.Image("../../../data/OpenCV-img/shape.png")
```

```python
kernel = np.ones((5, 5), np.uint8)
erosion = cv2.erode(img, kernel, iterations=1)
erosion2 = cv2.erode(img, kernel, iterations=2)  # 可以设置迭代次数

res = np.hstack((erosion, erosion2))
cv2.imwrite("../../../artifacts/OpenCV/03morphological_operations/erosion_shape.png", res)
```

```python
display.Image("../../../artifacts/OpenCV/03morphological_operations/erosion_shape.png")
```

## 膨胀

```python
kernel = np.ones((5, 5), np.uint8)
dilate = cv2.dilate(img, kernel, iterations=1)
dilate2 = cv2.dilate(img, kernel, iterations=2)

res = np.hstack((dilate, dilate2))
cv2.imwrite("../../../artifacts/OpenCV/03morphological_operations/dilate_shape.png", res)
```

```python
display.Image("../../../artifacts/OpenCV/03morphological_operations/dilate_shape.png")
```

## 开运算与闭运算
- 开运算：先腐蚀再膨胀
- 闭运算：先膨胀再腐蚀

```python
# 开运算
kernel = np.ones((5, 5), np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
cv2.imwrite("../../../artifacts/OpenCV/03morphological_operations/open_shape.png", opening)
```

```python
display.Image("../../../artifacts/OpenCV/03morphological_operations/open_shape.png")
```

```python
# 闭运算
kernel = np.ones((5, 5), np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
cv2.imwrite("../../../artifacts/OpenCV/03morphological_operations/close_shape.png", closing)
```

```python
display.Image("../../../artifacts/OpenCV/03morphological_operations/close_shape.png")
```

## 梯度运算
膨胀-腐蚀

```python
kernel = np.ones((3, 3), np.uint8)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
cv2.imwrite("../../../artifacts/OpenCV/03morphological_operations/gradient_shape.png", gradient)
```

```python
display.Image("../../../artifacts/OpenCV/03morphological_operations/gradient_shape.png")
```

## 顶帽与黑帽
- 顶帽：原图像-开运算
- 黑帽：闭运算-原图像

```python
# 顶帽
kernel = np.ones((3, 3), np.uint8)
tohat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
cv2.imwrite("../../../artifacts/OpenCV/03morphological_operations/tohat_shape.png", tohat)
```

```python
display.Image("../../../artifacts/OpenCV/03morphological_operations/tohat_shape.png")
```

```python
# 黑帽
kernel = np.ones((3, 3), np.uint8)
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
cv2.imwrite("../../../artifacts/OpenCV/03morphological_operations/blackhat_shape.png", blackhat)
```

```python
display.Image("../../../artifacts/OpenCV/03morphological_operations/blackhat_shape.png")
```
